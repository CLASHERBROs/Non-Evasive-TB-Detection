```python
import pandas as pd
import os
```

```python
filenames=os.listdir('CAPSTONE DATABASE/TB NEG/Rectified/')
# dataframe = pd.DataFrame()
tbneg1 = pd.read_csv("CAPSTONE DATABASE/TB NEG/Rectified/HC001.csv")
tbneg1.drop(tbneg1.columns[0:8], axis=1)
df1 = tbneg1.mean(axis = 0)
dataframe = pd.DataFrame(df1).transpose()
for i in range(len(filenames)-1):
    tbneg = pd.read_csv("CAPSTONE DATABASE/TB NEG/Rectified/"+filenames[i+1])
    tbneg.drop(tbneg.columns[0:8], axis=1, inplace=True)
    df = tbneg.mean(axis = 0)
    df_final = pd.DataFrame(df).transpose()
    dataframe.append(df_final)
    print(dataframe)
```
